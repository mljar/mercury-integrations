# Using only Vega-Altair

```python
# import packages 
import altair as alt 

# create data  
data = alt.Data(values=[{'x': 1, 'y': 8},
                        {'x': 9, 'y': 1},
                        {'x': 4, 'y': 9},
                        {'x': 3, 'y': 6},
                        {'x': 6, 'y': 2},
                        {'x': 5, 'y': 7},
                        {'x': 7, 'y': 4},
                        {'x': 2, 'y': 8},
                        {'x': 8, 'y': 3}])

# plot 
alt.Chart(data).mark_point().encode(
    x='x:Q',
    y='y:Q'
)
```

# Using Vega-Altair and Pandas

```python
# import packages 
import altair as alt 
import pandas as pd 

# create data 
df = pd.DataFrame({
    'x': [20,6,12,13,13,3,15,12,18,8],
    'y': [2,15,19,17,7,2,9,3,15,16]
})

# plot 
alt.Chart(df).mark_point().encode(
    x='x',
    y='y'
)
```
