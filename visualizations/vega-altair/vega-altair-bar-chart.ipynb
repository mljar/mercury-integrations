# Using only Vega-Altair

```python
# import packages 
import altair as alt 

# create data 
data = alt.Data(values=[{'x': 'A', 'y': 5},
                        {'x': 'B', 'y': 10},
                        {'x': 'C', 'y': 6},
                        {'x': 'D', 'y': 8},
                        {'x': 'E', 'y': 3},
                        {'x': 'F', 'y': 7},
                        {'x': 'G', 'y': 9},
                        {'x': 'H', 'y': 2},
                        {'x': 'I', 'y': 4}])

# plot 
alt.Chart(data).mark_bar().encode(
    x = 'x:N',
    y = 'y:Q'
).properties(
    width=500
)
```

# Using Vega-Altair and Pandas

```python
# import packages 
import altair as alt
import pandas as pd

# create data 
df = pd.DataFrame({
    'x': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    'y': [14, 9, 15, 8, 1, 14, 10, 9, 9]
})

# plot 
alt.Chart(df).mark_bar().encode(
    x='x',
    y='y'
).properties(
    width=500
)
```
