# Using only Matplotlib

```python
# import packages 
import matplotlib.pyplot as plt

# create data
x = [1, 2, 3, 4, 5]
y = [1.5, 5.3, 2.1, 6.2, 2.7]

# plot 
plt.plot(x,y)
plt.show()
```

# Using Matplotlib and Numpy 

```python
# import packages
import matplotlib.pyplot as plt
import numpy as np

# create data 
x = 1 + np.arange(8)
y = np.array([4.8, 5.5, 3.5, 6.6, 2.5, 4.6, 1.6, 3.9])

# plot 
plt.plot(x,y)
plt.show()
```

# Using Matplotlib and Pandas 

```python
# import packages
import matplotlib.pyplot as plt
import pandas as pd

# create data 
df = pd.DataFrame({'count': [1,2,3,4,5,6,7,8], 'value': [1.5,4.8,3.0,8.1,1.9,3.8,2.2,7.1]})

# plot 
plt.plot(df['count'], df['value'])
plt.show()
```
