```python
# import packages 
import matplotlib.pyplot as plt
import mercury as mr
```

```python
# mercury widget 
choice = mr.Checkbox(value=True, label="Show label")
```

```python
# create data
x = [20,30,50]
labels = ["Apples", "Bananas", "Oranges"] if choice.value else None

# plot
plt.pie(x, labels=labels)
plt.show()
```
