# Using only Matplotlib

```python
# import packages 
import matplotlib.pyplot as plt

# create data 
x = [1,2,3,4,5,6,7]
y = [11, 12, 13, 13, 11, 15, 12]

# plot 
plt.stem(x,y)
plt.show()
```

# Using Matplotlib and Numpy

```python
# import packages 
import matplotlib.pyplot as plt
import numpy as np

# create data 
x = 0.5 + np.arange(8)
y = np.array([4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0])

# plot
plt.stem(x,y)
plt.show()
```

# Using Matplotlib and Pandas

```python
# import packages
import matplotlib.pyplot as plt 
import pandas as pd

# create data 
df = pd.DataFrame({'numbers': [1,2,3,4,5,6,7], 'values': [12,8,7,14,10,11,8]})

# plot 
plt.stem(df['numbers'], df['values'])
plt.show()
```
