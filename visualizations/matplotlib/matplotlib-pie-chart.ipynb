# Using only Matplotlib

```python
# import packages
import matplotlib.pyplot as plt

# create data
x = [15,20,35,30]

# plot 
plt.pie(x)
plt.show()
```

# Using Matplotlib and Numpy

```python
# import packages 
import matplotlib.pyplot as plt
import numpy as np

# create data 
x = np.array([35, 25, 25, 15])

# plot 
plt.pie(x)
plt.show()
```

# Using Matplotlib and Pandas 

```python
# import packages
import matplotlib.pyplot as plt
import pandas as pd 

# create data 
df = pd.DataFrame({'percent': [15,25,35,25]})

# plot 
plt.pie(df['percent'])
plt.show()
```
