```python
# import packages 
import seaborn as sns 
import mercury as mr 
```

```python
# mercury widget 
selection = mr.Select(label="Choose Data:", value="Plot 1", choices=["Plot 1", "Plot 2", "Plot 3"])
```

```python
# create data 
x1 = [4, 10, 15, 8, 13, 11, 19, 9, 14, 20]
x2 = [18, 11, 10, 16, 15, 20, 9, 15, 15, 3]
x3 = [15, 11, 13, 9, 4, 9, 17, 7, 5, 8]

# plot 
if "Plot 1" in selection.value: 
    _ = sns.lineplot(x1)
elif "Plot 2" in selection.value:
    _ = sns.lineplot(x2)
elif "Plot 3" in selection.value:
    _ = sns.lineplot(x3)
```
