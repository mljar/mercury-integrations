```python
# import packages 
import seaborn as sns 
import mercury as mr
```

```python
# mercury widget
slider = mr.Slider(value=10, min=8, max=20, label="How many dots?")
```

```python
# create data
all_x = [22, 6, 17, 13, 10, 3, 17, 21, 22, 22, 9, 11, 10, 12, 16, 16, 4, 6, 14, 9]
x = all_x[0:slider.value]

# plot 
_ = sns.scatterplot(x)
```
