# Using only Seaborn

```python
# import packages
import seaborn as sns 

# create data 
x = [4, 10, 15, 8, 13, 11, 19, 9, 14, 20]

# plot 
_ = sns.lineplot(x)
```

# Using Seaborn and Numpy

```python
# import packages 
import seaborn as sns 
import numpy as np 

# create data 
x = np.array([18, 11, 10, 16, 15, 20, 9, 15, 15, 3])

# plot 
_ = sns.lineplot(x)
```

# Using Seaborn and Pandas

```python
# import packages 
import seaborn as sns 
import pandas as pd 

# create data
df = pd.DataFrame({'value': [15, 11, 13, 9, 4, 9, 17, 7, 5, 8]})

# plot 
_ = sns.lineplot(df)
```
