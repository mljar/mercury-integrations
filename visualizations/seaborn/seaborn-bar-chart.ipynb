# Using only Seaborn

```python
# import packages 
import seaborn as sns

# create data 
y = [1.5,3.8,5.7,7.1,2.9]

# plot 
_ = sns.barplot(y)
```

# Using Seaborn and Numpy

```python
# import packages 
import seaborn as sns
import numpy as np

# create data 
y = np.array([3.6,5.4,2.1,6.9,1.9])

# plot 
_ = sns.barplot(y)
```

# Using Seaborn and Pandas

```python
# import packages 
import seaborn as sns
import pandas as pd 

# create data 
df = pd.DataFrame({'count': [1,2,3,4,5], 'value': [3.4,5.1,6.7,4.3,8.1]})

# plot 
_ = sns.barplot(df, x='count',y='value')
```
