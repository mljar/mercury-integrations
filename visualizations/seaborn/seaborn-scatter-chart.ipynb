# Using only Seaborn

```python
# import packages 
import seaborn as sns

# create data 
x = [3, 9, 9, 7, 2, 10, 9, 6, 6, 10]

# plot 
_ = sns.scatterplot(x)
```

# Using Seaborn and Numpy 

```python
# import packages 
import seaborn as sns 
import numpy as np 

# create data 
x = np.array([3, 9, 7, 4, 3, 1, 10, 7, 10, 6])

# plot 
_ = sns.scatterplot(x)
```

# Using Seaborn and Pandas

```python
# import packages 
import seaborn as sns 
import pandas as pd 

# create data 
df = pd.DataFrame({ 'value': [4, 3, 6, 4, 1, 5, 10, 2, 1, 3]})

# plot 
_ = sns.scatterplot(df)
```
